```python
import random
```

```python
class Functions:
    @staticmethod
    def shuffle(df):
        dl = df.values.tolist()
        length = len(dl)
        for x in range(length):
            i = random.randint(0, length-x)
            dl[i], dl[-1-x] = dl[-1-x], dl[i]
        new_df = pd.DataFrame(data=dl, columns=df.columns)
        return new_df

    @staticmethod
    def NormalizeData(df):
        #przeprowadzenie normalizacji danych
        mean_vals = df.mean()
        std_vals = df.std()
        normalized_data = (df - mean_vals) / std_vals
        return normalized_data
```

```python
import numpy as np
from sklearn.model_selection import KFold

class SoftSet:
    def __init__(self, n_iters=1000, lambda_param=0.01):
        self.n_iters = n_iters
        self.lambda_param = lambda_param

    def fit(self, X, Y):
        self.X = X
        self.Y = np.zeros(len(X[0]))
        num_features = len(X[0]) if len(X) > 0 else 0
        for _ in range(self.n_iters):
            for i, (x_i, y_i) in enumerate(zip(X, Y)):
                if y_i * self.classify(x_i) <= 1:
                    self.Y += y_i * x_i - 2 * self.lambda_param * self.Y

    def predict(self, X):
        results = []
        for x_i in X:
            classification = self.classify(x_i)
            results.append(np.sign(classification))
        return results

    def classify(self, sample):
        return np.dot(sample, self.Y)
```

```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.naive_bayes import GaussianNB

class Bayes:
    def __init__(self):
        self.model = GaussianNB()  # Użycie modelu Gaussian Naive Bayes

    def fit(self, X, Y):
        self.model.fit(X, Y)  # Dopasowanie danych do modelu Bayesa

    def predict(self, X):
        return self.model.predict(X)  # Wykonanie prognozy za pomocą modelu Bayesa
```

```python
import numpy as np
from collections import Counter
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

class KNN:
    def __init__(self, k=3, m=2):
        self.k = k
        self.m = m

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        predictions = []
        for test_record in X_test:
            distances = []
            for train_record in self.X_train:
                distances.append(self.distance(test_record, train_record))
            k_nearest_indices = np.argsort(distances)[:self.k]
            k_nearest_labels = [self.y_train[i] for i in k_nearest_indices]
            most_common_label = Counter(k_nearest_labels).most_common(1)[0][0]
            predictions.append(most_common_label)
        return predictions

    def distance(self, record1, record2):
        return np.linalg.norm(record1 - record2, ord=self.m)

    def score(self, X_test, y_test):
        y_pred = self.predict(X_test)
        return accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix

# Przygotowanie danych
# 'df' to DataFrame zawierający nasze dane
# 'output' jest kolumną, którą chcemy przewidzieć

#Dla danych znormalizowanych
df_norm=Functions.NormalizeData(serce)
# Podział danych na cechy (X) i etykietę (y)
X_norm = df_norm[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh','exng','oldpeak', 'slp', 'thall', 'caa']]
y_norm = df_norm['output']

# Podział danych na zestaw treningowy i testowy
X_train_norm, X_test_norm, y_train_norm, y_test_norm = train_test_split(X_norm, y_norm, test_size=0.35, random_state=42)
y_test_norm=np.sign(y_test_norm)

model2 = SoftSet()
model2.fit(X_train_norm.values, y_train_norm)

# Przewidywanie klas dla danych testowych
y_pred_norm = model2.predict(X_test_norm.values)

# Oblicz macierz pomyłek dla modelu SoftSetSVM
conf_matrix_norm = confusion_matrix(y_test_norm.values, y_pred_norm)

# Wyświetl macierz pomyłek za pomocą seaborn
plt.figure(figsize=(2, 2))
sns.heatmap(conf_matrix_norm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for SoftSet Model")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
# Ocena wydajności modelu
accuracy = accuracy_score(y_test_norm, y_pred_norm)
print("Accuracy when model is train with normalized soft set:", accuracy*100)

# Raport klasyfikacji
print(classification_report(y_test_norm, y_pred_norm, zero_division=1))

print('\n###########################################################\n')

#Dla danych znormalizowanych i wymieszanych
df_norms=Functions.NormalizeData(serce)
df_norms=Functions.shuffle(df_norms)
# Podział danych na cechy (X) i etykietę (y)
X_norms = df_norms[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh','exng','oldpeak', 'slp', 'thall', 'caa']]
y_norms = df_norms['output']

# Podział danych na zestaw treningowy i testowy
X_train_norms, X_test_norms, y_train_norms, y_test_norms = train_test_split(X_norms, y_norms, test_size=0.35, random_state=42)

y_test_norms=np.sign(y_test_norms)

model3 = SoftSet()
model3.fit(X_train_norms.values, y_train_norms)

# Przewidywanie klas dla danych testowych
y_pred_norms = model3.predict(X_test_norms.values)

# Oblicz macierz pomyłek dla modelu SoftSetSVM
conf_matrix_norms = confusion_matrix(y_test_norms, y_pred_norms)

# Wyświetl macierz pomyłek za pomocą seaborn
plt.figure(figsize=(2, 2))
sns.heatmap(conf_matrix_norms, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for SoftSet Model")
plt.ylabel("True Label")
plt.show()

# Ocena wydajności modelu
accuracy = accuracy_score(y_test_norms, y_pred_norms)
print("Accuracy when model is train with normalized and shuflled soft set:", accuracy*100)

# Raport klasyfikacji
print(classification_report(y_test_norms, y_pred_norms, zero_division=1))
print('\n###########################################################\n')

#Dla danych znormalizowanych
df_normb=Functions.NormalizeData(serce)
# Podział danych na cechy (X) i etykietę (y)
X_b = df_normb[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh','exng','oldpeak', 'slp', 'thall', 'caa']]
y_b = df_normb['output']
y_b = np.sign(df_normb['output'])
# Podział danych na zestaw treningowy i testowy
X_trainb, X_testb, y_trainb, y_testb = train_test_split(X_b, y_b, test_size=0.35, random_state=42)
y_testb=np.sign(y_testb)
# Inicjalizacja i dopasowanie modelu SoftSetSVM
model4 = Bayes()
model4.fit(X_trainb, y_trainb)

# Przewidywanie klas dla danych testowych
y_pred_b = model4.predict(X_testb.values)

# Oblicz macierz pomyłek dla modelu SoftSetSVM
conf_matrix_b = confusion_matrix(y_testb.values, y_pred_b)

# Wyświetl macierz pomyłek za pomocą seaborn
plt.figure(figsize=(2, 2))
sns.heatmap(conf_matrix_b, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for Bayes Model")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Ocena wydajności modelu
accuracyb = accuracy_score(y_testb, y_pred_b)
print("Accuracy when model is train with bayes:", accuracyb*100)

# Raport klasyfikacji
print(classification_report(y_testb, y_pred_b, zero_division=1))

print('\n###########################################################\n')
# Dla danych znormalizowanych
df_normb = Functions.NormalizeData(serce)

# Podział danych na cechy (X) i etykietę (y)
X_b = df_normb[['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh','exng','oldpeak', 'slp', 'thall', 'caa']]
y_b = df_normb['output']

# Podział danych na zestaw treningowy i testowy
X_trainb, X_testb, y_trainb, y_testb = train_test_split(X_b, y_b, test_size=0.35, random_state=42)

# Konwersja do macierzy numpy
X_trainb = X_trainb.values
X_testb = X_testb.values
y_trainb = y_trainb.values.astype(int)  # Konwersja etykiet do typu całkowitoliczbowego
y_testb = y_testb.values.astype(int)    # Konwersja etykiet do typu całkowitoliczbowego

# Inicjalizacja i dopasowanie modelu KNN
model_knn = KNN(k=5, m=2)  # Ustawienie liczby sąsiadów na 5, metryka Minkowskiego z parametrem m=2 (domyślnie euklidesowa)
model_knn.fit(X_trainb, y_trainb)

# Przewidywanie klas dla danych testowych
y_pred_knn = model_knn.predict(X_testb)

# Obliczenie macierzy pomyłek dla modelu KNN
conf_matrix_knn = confusion_matrix(y_testb, y_pred_knn)

# Wyświetlenie macierzy pomyłek za pomocą seaborn
plt.figure(figsize=(2, 2))
sns.heatmap(conf_matrix_knn, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for KNN Model")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Ocena wydajności modelu
accuracy_knn = model_knn.score(X_testb, y_testb)
print("Accuracy when model is trained with KNN:", accuracy_knn * 100)

# Raport klasyfikacji
print(classification_report(y_testb, y_pred_knn, zero_division=1))
```
